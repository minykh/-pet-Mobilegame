# Проект: вариант 1

## Задание 3

Для качественной аналитики события предлагаю посмотреть следующие метрики:  
* Retention Rate - процент возврата игроков во время ивента
* DAU и MAU(если ивент длится больше месяца), количество уникальных пользователей за день и месяц соответственно.
* ASD — среднее проведенное игроком время.
* ARPU и ARPPU - если у нас есть платные опции, такие как боевой пропуск, лотереи за реальные деньги и прочее.
* Также я бы создал воронку имеющихся уровней, чтобы понять какой процент игроков до какого уровня дошел.

При введении усложненной механики желательно провести А/Б-тест (например на разных регионах), и обратить особенное внимание на показатель Retention, а также сравнить показатели созданной ранее воронки.
